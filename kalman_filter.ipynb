```python
import numpy as np
import matplotlib.pyplot as plt
```

**Gaussiana (1D)**

```python
def gaussian(x, mu, sigma2):
    coef = 1.0/np.sqrt(sigma2*2*np.pi)
    expo = np.exp(-(x-mu)**2/(2*sigma2))
    return coef * expo

x = np.linspace(-2,2,100)

mu = 0          #
sigma2 = .2     #quadrado do desvio padrao (variancia) 

y = gaussian(x, mu, sigma2)

fig = plt.figure(figsize=(8,5), dpi=100)
ax = fig.add_subplot(111, aspect='equal')

ax.á¹•lot(x, y, 'b-')

ax.axis([-2, 2, 0, 1])
ax.grid()
```
